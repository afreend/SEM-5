# catagorical data

```python
Salary_Data=pd.DataFrame({"Salary":[10000,20000,30000,40000,50000,60000],"Range":['low','low','mid','mid','high','high']})
```

```python
Salary_Data
```

```python
code=Salary_Data.Range.map({'low':3,'high':1,'mid':2})
```

```python
code
```

```python
publish=pd.DataFrame(Salary_Data['Salary'],columns=['Salary'])
```

```python
publish["code"]=code
```
